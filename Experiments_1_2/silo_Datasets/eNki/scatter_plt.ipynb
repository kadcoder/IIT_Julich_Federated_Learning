```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the post harmo data
df_post = pd.read_csv('/home/tanurima/germany/brain_age_parcels/kunal/Filtered_Datasets/eNki/Test_eNki_m1_150.csv')

# Create scatter plot for post harmo
plt.figure(figsize=(8, 6))
plt.scatter(df_post['Actual_Age'], df_post['Predicted_Age'], color='blue', alpha=0.7)
plt.title('Post Harmo m1(eNki): Actual Age vs Predicted Age')
plt.xlabel('Actual Age')
plt.ylabel('Predicted Age')
plt.grid(True)
plt.show()
plt.savefig('/home/tanurima/germany/brain_age_parcels/kunal/Filtered_Datasets/eNki/Test_eNki_m1.png')
```

```python
from sklearn.metrics import r2_score
# Calculate R² for Post Harmo
r2_post = r2_score(df_post['Actual_Age'], df_post['Predicted_Age'])
print(f'R² value m1 (Post Harmo): {r2_post:.4f}')
```

```python
# Calculate Pearson correlation for Post Harmo
corr_post = np.corrcoef(df_post['Actual_Age'], df_post['Predicted_Age'])[0, 1]
print(f'Pearson Correlation Coefficient m1 (Post Harmo): {corr_post:.4f}')
```
