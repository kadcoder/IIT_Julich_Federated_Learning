```python
import matplotlib.pyplot as plt,os
import matplotlib.image as mpimg
```

```python
# Load the images
path = '/home/tanurima/germany/brain_age_parcels/kunal/Filtered_Datasets'
img1 = mpimg.imread(os.path.join(path,'tsne-silos-tr-distribution.png'))
img2 = mpimg.imread(os.path.join(path,'tsne-silos-harmo-tr-distribution.png'))
```

```python
# Create a figure with subplots
plt.figure(figsize=(12, 6))

# Plot the first image
plt.subplot(1, 2, 1)
plt.imshow(img1)
plt.axis('off')
plt.title('tsne-distribution (Pre-H)')

# Plot the second image
plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.axis('off')
plt.title('tsne-distribution (Post-H)')

# Display the combined plot
plt.tight_layout()
plt.show()
```
